```python
class LegoBrickFitter:
    def __init__(self, arr):
        arr[arr != 0] = 1
        self.arr = arr.copy()
        self.arr_byLegoType = arr.copy().astype(str)
        
        self.bricks_dims = {f'{k[0]}x{k[1]}': k for k in [[2,4], [2,2], [2,1], [1,1]]}
        self.bricks_count_byType = {f'{k[0]}x{k[1]}': 0 for k in [[2,4], [2,2], [2,1], [1,1]]}
        self.bricks_count_total = 2

    def fit_larger_bricks(self):
        for i, row in enumerate(self.arr):
            if i % 2 != 0:
                continue
            for j, pixel in enumerate(row):
                if pixel != 1:
                    continue
                self._try_fitting_brick(i, j)

    def _try_fitting_brick(self, i, j):
        for brick_hw in [[2,4], [2,2], [2,1]]:
            temp = self.arr[i:i+brick_hw[0], j:j+brick_hw[1]]
            if temp.shape == tuple(brick_hw) and np.all(temp == 1): 
                self._place_brick(i, j, brick_hw)
                break

    def _place_brick(self, i, j, brick_hw):
        self.arr[i:i+brick_hw[0], j:j+brick_hw[1]] = self.bricks_count_total
        self.arr_byLegoType[i:i+brick_hw[0], j:j+brick_hw[1]] = f'{brick_hw[0]}x{brick_hw[1]}'
        
        brick_name = [k for k, v in self.bricks_dims.items() if v == brick_hw][0]
        self.bricks_count_byType[brick_name] += 1
        self.bricks_count_total += 1

    def fit_smallest_bricks(self):
        for i, row in enumerate(self.arr):
            for j, pixel in enumerate(row):
                if pixel == 1:
                    self.arr[i, j] = self.bricks_count_total
                    self.arr_byLegoType[i, j] = '1x1'
                    self.bricks_count_byType['1x1'] += 1
                    self.bricks_count_total += 1

    def fit_all_bricks(self):
        self.fit_larger_bricks()
        self.fit_smallest_bricks()
```

```python
layers_list = ['layer2_shallowSea', 'layer2_land', 
               'layer3_hk_1', 'layer3_hk_2', 'layer3_hk_3', 'layer3_hk_4', 
               'layer4_built_1', 'layer4_built_2', 'layer4_built_3']

layers_dict = {}
for name in layers_list: 
    layers_dict[name] = {}
    arr = np.load(f'data/{name}.npy')
    fitter = LegoBrickFitter(arr)
    fitter.fit_all_bricks()
    layers_dict[name]['array'] = fitter.arr
    layers_dict[name]['array_byType'] = fitter.arr_byLegoType
    layers_dict[name]['bricks_count_byType'] = fitter.bricks_count_byType
    layers_dict[name]['bricks_count_total'] = fitter.bricks_count_total
```

```python
layers_df = pd.DataFrame({
    'layer_name': layers_dict.keys(), 
    'layer_color': ['light blue', 'gray', 
                    'dark green', 'green', 'light green', 'yellow', 
                    'orange', 'orange', 'orange']
})

def brick_counts(row): 
    name = row.layer_name
    for brick_type in ['2x4', '2x2', '2x1', '1x1']: 
        row[brick_type] = layers_dict[name]['bricks_count_byType'][brick_type]
    row['total'] = layers_dict[name]['bricks_count_total']
    return row 
    
layers_df = layers_df.apply(lambda row: brick_counts(row), axis=1)
layers_df
```

```python
layers_list = ['layer2_shallowSea', 'layer2_land', 
               'layer3_hk_1', 'layer3_hk_2', 'layer3_hk_3', 'layer3_hk_4', 
               'layer4_built_1', 'layer4_built_2', 'layer4_built_3']
brickTypes = ['2x4', '2x2', '2x1', '1x1']

for layer in layers_list: 
    print(layer)
    for i, brickType in enumerate(brickTypes): 
        arr = layers_dict[layer]['array_byType'].copy()
        print(brickType)
    
        # Create a mask for each color
        white_mask = arr == '0'
        black_mask = arr == brickType
        grey_mask = np.isin(arr, brickTypes[:i])  # bricks before the current brickType
    
        arr_new = np.zeros_like(arr, dtype=int)
        
        arr_new[white_mask] = 0
        arr_new[grey_mask] = 1
        arr_new[black_mask] = 2
        
        fig, ax = plt.subplots(figsize=(10, 10))
        ax.set_title(f'{layer}: {brickType}')
        ax.set_aspect('equal', 'box')
        
        # make colormap (bwg)
        cmap = ListedColormap(['white', 'darkgrey', 'black'])
        ax.imshow(arr_new, cmap=cmap, origin='lower', aspect='equal')
        
        # add major and minor ticks
        ax.set_xticks(np.arange(-0.5, arr.shape[1], 1), minor=True)
        ax.set_yticks(np.arange(-0.5, arr.shape[0], 1), minor=True)
        ax.grid(which='minor', color='gray', linestyle='-', linewidth=0.5)
        ax.set_xticks(np.arange(-0.5, arr.shape[1], 32), minor=False)
        ax.set_yticks(np.arange(-0.5, arr.shape[0], 32), minor=False)
        ax.grid(which='major', color='gray', linestyle='-', linewidth=1.5)
        
        # show
        fig.savefig(f"results/{layer}_{brickType}.png", dpi=300)
        plt.show()
```

```python
def plot_map_all(name, arr, save=False, plot=True): 
    fig, ax = plt.subplots(figsize=(10, 10))
    ax.set_title(f'{name}: all')
    ax.set_aspect('equal', 'box')
    
    # make colormap (bw)
    unique_values = np.unique(arr)
    cmap = ListedColormap(['white' if value == 0 else 'black' for value in unique_values])
    ax.imshow(arr, cmap=cmap, origin='lower', aspect='equal')
    
    # add major and minor ticks
    ax.set_xticks(np.arange(-0.5, arr.shape[1], 1), minor=True)
    ax.set_yticks(np.arange(-0.5, arr.shape[0], 1), minor=True)
    ax.grid(which='minor', color='gray', linestyle='-', linewidth=0.5)
    ax.set_xticks(np.arange(-0.5, arr.shape[1], 32), minor=False)
    ax.set_yticks(np.arange(-0.5, arr.shape[0], 32), minor=False)
    ax.grid(which='major', color='gray', linestyle='-', linewidth=1.5)
    
    # Save and show figure
    if save: 
        fig.savefig(f"results/{name}_all.png", dpi=300)
    if plot: 
        print(name)
        plt.show()

for name, info_dict in layers_dict.items(): 
    arr = info_dict['array']
    plot_map_all(name, arr, save=True, plot=True)
```

# Counting all pieces 

```python
layers_list = ['layer2_shallowSea', 'layer2_land', 
               'layer3_hk_1', 'layer3_hk_2', 'layer3_hk_3', 'layer3_hk_4', 
               'layer4_built_1', 'layer4_built_2', 'layer4_built_3']

counts_dict = {}
for layer in layers_list: 
    counts_dict[layer] = {}

    arr_plot = layers_dict[layer]['array'].copy()
    plot_map_all(layer, arr_plot, save=False, plot=True)

    arr = layers_dict[layer]['array_byType'].copy()
    unique_values, counts = np.unique(arr, return_counts=True)

    brickCount_pixel = dict(zip(unique_values[1:], counts[1:]))
    for brickName, pixelCount in brickCount_pixel.items(): 
        height = int(brickName[0])
        width = int(brickName[-1])
        area = height * width 
        brickCount = pixelCount / area 
        counts_dict[layer][brickName] = int(brickCount)
```

```python
df = pd.DataFrame(counts_dict).T # .drop(columns=['0'])
df = df.fillna(0)
def count_total(row): 
    row['total'] = sum(row)
    return row
df = df.apply(lambda row: count_total(row), axis=1)
df = df.reset_index(names='layer_name')
df['layer_color'] = pd.Series(['blue', 'light gray', 
                    'green', 'bright green', 'lime', 'yellow', 
                    'orange', 'orange', 'orange'])
print(df.total.sum())
df = df[['layer_name', 'layer_color', '1x1', '2x1', '2x4', '2x2', 'total']]

col_dict = {
    '1x1': '1x1_3024', 
    '2x1': '1x2_3023', 
    '2x2': '2x2_3022', 
    '2x4': '2x4_3020'
}
df.rename(columns=col_dict, inplace=True)

df.to_csv('results/brick_count_all.csv')
df
```
